```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.preprocessing import image
```

```python
# Load the saved model
loaded_model = tf.keras.models.load_model('/Users/akshay/projects/Gender_classification/gender_classification_model.h5')

# Specify the paths to the images to be predicted
paths = ["/Users/akshay/projects/Gender_classification/Dataset/Test/Male/161496.jpg",] #you can also add additional image paths for prediction


# Iterate over the paths and predict the gender for each image
for path in paths:
    # Load the image and resize it to the required input shape (64x64)
    img = image.load_img(path, target_size=(64, 64))

    # Convert the image to a numpy array
    x = image.img_to_array(img)

    # Add an extra dimension to the array to match the model's input shape
    x = np.expand_dims(x, axis=0)

    # Stack the array vertically to create a batch of one image
    images = np.vstack([x])

    # Use the loaded model to predict the class probabilities for the image
    classes = loaded_model.predict(images, batch_size=1)

    # Print the path of the image
    print("Image:", path)

    # Print the predicted class probabilities
    print("Predicted class probabilities:", classes[0])

    # Check if the predicted probability is greater than 0.5 (threshold for classification)
    if classes[0] > 0.5:
        print("The image is predicted to be of a man.")
    else:
        print("The image is predicted to be of a woman.")

    # Display the image
    plt.imshow(img)
    plt.show()
```
